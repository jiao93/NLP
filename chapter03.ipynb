## 20. Reading JSON data

```python
import json
f = open('jawiki-country.json')
data = [json.loads(i) for i in f.readlines()]
```

## 22. Extracting category from lines

```python
import re

for i in data:
    g = re.search(r'\[\[Category:(.*)\|.*\]\]',i['text'])
    if g:
        print(g.group(0))
        print(g.group(1))
```

## 23. Section Structure

```python
import re

for i in data:
        m = re.findall(r'(={2,})(.*?)={2,}',i['text'])
        for g in m:
            print(g[1].strip())
            print(len(g[0])-1)
```

## 24. Extracting file reference

```python
for i in data:
        m = re.findall(r'\[\[ファイル:([^\]]*?)\|.*\]\]',i['text'])
        for g in m:
            print(g)
```

## 25. Extracting basic infomation template

```python
for i in data:
        l=i['text']
        g = re.search(r'\{\{基礎情報(.|\n)*?(\n|\|)\}\}',l)
        if g:
            d = dict()
            for j in g.group(0).splitlines()[1:-1]:
                equal_sign_index = j.find('=')
                k = j[:equal_sign_index].strip('|').strip()
                v = j[equal_sign_index+1:].strip('|').strip()
                d[k]=v
            i['basic_info']=d
```

## 26. Clearing emphasis markup

```python
for i in data:
    if 'basic_info' in i:
        for k,v in i['basic_info'].items():
            s = re.sub(r"'''''(?P<c>.+?)'''''",r'\g<c>',v)
            s = re.sub(r"\'{2,3}(?P<c>.+?)\'{2,3}",r'\g<c>',s)
            i['basic_info'][k]=s
```

## 27. Clearing internal link

```python
for i in data:
    if 'basic_info' in i:
        for k,v in i['basic_info'].items():
            s = re.sub(r"\[\[.+\|(?P<c>.+?)\]\]",r'\g<c>',v)
            s = re.sub(r"\[\[(?P<c>.+?)\]\]",r'\g<c>',s)
            i['basic_info'][k]=s
```

## 28. Clearing more markup

```python
for i in data[:20]:
    if 'basic_info' in i:
        for k,v in i['basic_info'].items():
            s = re.sub(r"<ref.*/>",'',v)
            i['basic_info'][k]=s
        
```

## 29. Get flag URL

```python
import requests
def get_image_url(filename):
    r = requests.get(r'https://commons.wikimedia.org/w/api.php',params={'action':'query','format':'json',
                                                                       'titles':'file:'+filename,'prop':'imageinfo','iiprop':'url'})
    result_json = r.json()
    page = result_json['query']['pages'].popitem()[1]
    return page['imageinfo'][0]['url']

url = get_image_url(data[0]['basic_info']['国旗画像'])
print(url)
```
